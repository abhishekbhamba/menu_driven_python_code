```python
# write a programe to make menu driven programe.
import pickle, os

def save():
    book={}
    f=open("flip.dat","ab")
    bcode=int(input("enter book code: "))
    topic=input("enter subject of the book: ")
    aut=input("enter author name of the subject: ")
    price=float(input("enter price of the book: "))
    book[bcode]=[topic,aut,price]
    pickle.dump(book,f)
    print("data saved...")
    f.close()
    
def display():
    f=open("flip.dat","rb")
    try:
        while(True):
            x=pickle.load(f)
            for i in x:
                print(i,end="\t")
                for j in x[i]:
                    print(j,end="\t")
                print()
    except:
        f.close()
        
def search():
    f=open("flip.dat","rb")
    bcode=int(input("enter book code to search: "))
    flag=0
    try:
        while(True):
            x=pickle.load(f)
            for i in x:
                if(i==bcode):
                    print(i,end="\t")
                    for j in x[i]:
                        print(j,end="\t")
                        flag=1
                    print()
    except:
        f.close()
    if(flag==0):
        print("not found")
        
def delete():
    f=open("flip.dat","rb")
    f1=open("temp.dat","wb")
    bcode=int(input("enter bcode to delete: "))
    flag=0
    try:
        while(True):
            x=pickle.load(f)
            for i in x:
                if(i!=bcode):
                    pickle.dump(x,f1)
            else:
                flag=1
    except:
        f.close()
        f1.close()
    if(flag==0):
        print("not deleted")
    else:
        print("sucessfully deleted....")
    os.remove("flip.dat")
    os.rename("temp.dat","flip.dat")
    
def update():
    f=open("flip.dat","rb")
    f1=open("temp.dat","wb")
    d={}
    flag=0
    bcode=int(input("enter bcode to update: "))
    try:
        while(True):
            x=pickle.load(f)
            for i in x:
                if(i==bcode):
                    p=float(input("enter new price: "))
                    l=[x[i][0],x[i][1],p]
                    d[bcode]=l
                    pickle.dump(d,f1)
                    flag+=1
                else:
                    pickle.dump(x,f1)
    except:
        f.close()
        f1.close()
    if(flag==0):
        print("not updated")
    else:
        print(flag,"updated...")
    os.remove("flip.dat")
    os.rename("temp.dat","flip.dat")
                    
    
print("1.SAVE\n2.DISPLAY\n3.SEARCH\n4.DELETE\n5.UPDATE")
ch="y"
while(ch.lower()=="y"):
    op=int(input("enter option: "))
    if(op==1):
        save()
    elif(op==2):
        display()
    elif(op==3):
        search()
    elif(op==4):
        delete()
    elif(op==5):
        update()
    else:
        print("invalid")
    ch=input("enter to continue y/n? ")
```
