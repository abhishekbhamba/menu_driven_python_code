```python
# save, display, delete, cost, search, update?

import pickle,os
def save():
    f=open("badal.dat","ab")
    bcode=int(input("enter book code: "))
    topic=input("enter subject: ")
    author=input("enter author of the book: ")
    price=float(input("enter price of the book: "))
    l=[bcode,topic,author,price]
    pickle.dump(l,f)
    print("data saved......")
    f.close()
    
def display():
    print("=========================")
    print("BCODE\tTOPIC\tAUTHOR\tPRICE")
    print("=========================")
    f=open("badal.dat","rb")
    flag=0
    try:
        while(True):
            x=pickle.load(f)
            for i in x:
                print(i,end="\t")
            print()
    except:
        f.close()
        
def delete():
    f=open("badal.dat","rb")
    f1=open("temp.dat","wb")
    flag=0
    bcode=int(input("enter book code: "))
    try:
        while(True):
            x=pickle.load(f)
            if(x[0]!=bcode):
                pickle.dump(x,f1)
            else:
                flag=1
    except:
        f.close()
        f1.close()
    if(flag==0):
        print("not deleted / wrong data entered")
    else:
        print("deleted succesfully......")
    os.remove("badal.dat")
    os.rename("temp.dat","badal.dat")
    
def cost():
    f=open("badal.dat","rb")
    f1=open("costly.dat","wb")
    flag=0
    try:
        while(True):
            x=pickle.load(f)
            if(x[3]>=500):
                flag+=1
                pickle.dump(x,f1)
                print()
    except:
        f.close()
        f1.close()
    if(flag==0):
        print(flag,"found")
    else:
        print(flag,"succesfully....")
        
def search():
    f=open("badal.dat","rb")
    topic=input("enter subject to search: ")
    flag=0
    try:
        while(True):
            x=pickle.load(f)
            if(x[1].lower()==topic.lower()):
                flag=1
                for i in x:
                    print(i,end="\t")
                print()
    except:
        f.close()
    if(flag==0):
        print("not found")
        
def update():
    f=open("badal.dat","rb")
    f1=open("update.dat","wb")
    flag=0
    bcode=int(input("enter bcode to update: "))
    try:
        while(True):
            x=pickle.load(f)
            if(x[0]==bcode):
                au=input("enter new author: ")
                p=int(input("enter new price: "))
                l=[x[0],x[1],au,p]
                pickle.dump(l,f1)
                flag=1
            else:
                pickle.dump(x,f1)
    except:
        f.close()
        f1.close()
    if(flag==0):
        print("not found")
    else:
        print("update sucessfully...")
    os.remove("badal.dat")
    os.rename("update.dat","badal.dat")
            
            
            
print("1.save\n2.display\n3.delete\n4.cost\n5.search\n6.update")
ch="y"
while(ch.lower()=="y"):
    op=int(input("enter option: "))
    if(op==1):
        save()
    elif(op==2):
        display()
    elif(op==3):
        delete()
    elif(op==4):
        cost()
    elif(op==5):
        search()
    elif(op==6):
        update()
    else:
        print("invalid no. enter")
    ch=input("enter y/n to continue")
```
